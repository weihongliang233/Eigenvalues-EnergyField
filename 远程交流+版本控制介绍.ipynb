# 动机

我花了些许时间+金钱部署了这个带jupyter的服务器。简单来讲，就是你可以随时随地打开浏览器访问这个网址，从而进入jupyter。你每次按下shift+enter的时候都发生了这么一件事：你的浏览器将当前的代码发送给我的服务器；服务器上面的内核进行运算，再把运算结果传回来。

做这件事的主要原因是与人合作的过程中经常要发生文件交换；这其实是很费力的一件事情。也许不经常计算机的人会觉得这没啥，我非常理解这种心情，毕竟不是什么非常麻烦的事情，自己进行人肉版本控制也可以接受。在我还没接触计算机的时候我也抱有同样的想法。但是随着协作次数越来越多，一套方便的工作流也确实值得我们去花费一点时间。而且我非常有信心让你们在十分钟之内就能知道这套工作流是怎么回事。

## 基本操作

就像在MMA里面一样，你想要运行某个单元的话，选中这个单元然后按下shift+enter键即可。

```
Plot[Sin[x+1],{x,0,1}]
```

如果你想新建单元格的话，只需按下esc键，再按下b键，即可在当前单元下面新建一个单元。

每次你工作完以后，想要把代码分享给我，只需要保存该文件，然后告诉我你写好了。这时我再自己打开这个网址，就能看见你的工作，并且做出修改。

左边这一栏可以让你选择文件路径。你可以新建或者删除或者进行别的操作，都没问题。

## 选择内核

jupyter有个好处就是可以自由选择内核。也就是说你可以在这里使用各种各样的语言（不止MMA）。你只需要去到上方的"Kernel"一栏，下拉，选择“Change Kernel”即可选择你想要的内核。不过我想你一般都会采用Wolfram内核。当前内核可以在右上角看到。如果当前内核不是wolfram的话，直接按照刚刚的步骤选择内核即可。
